```python
#Entrar a una página y obtener información

#Importar librerías
import os
import pandas as pd
import time
from bs4 import BeautifulSoup
import requests
```

```python
os.chdir('H:/Mi unidad/economía/medicamentos')
```

```python
#Parámetros para la página
url_basica="https://www.vademecum.es/atc-mx-L01"
clave=["A","B","C","D","E","F","G","H","J","K","L","M","N","P","Q","R","S","T","V","W","X","Y","Z"]
#usar headers para que no nos detecte como un bot
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
```

```python
#Extraer información
datos=[]
for i in clave:
    for j in clave:
        url=url_basica+i+j
        r = requests.get(url, headers=headers)
        soup = BeautifulSoup(r.text, 'html.parser')
        for k in soup.find_all('a', class_='button atc'):
            datos.append([k.text])
        time.sleep(2)      
```

```python
datos
```

```python
#Crear un dataframe y separar la columna clave
df=pd.DataFrame(datos,columns=["clave"])
#Split de la columna clave
df[["clave","nombre"]]=df["clave"].str.split("- ",expand=True)
#Eliminar espacios en blanco
df["clave"]=df["clave"].str.strip()
print("Número total de registros: ",len(df))
df
```

```python
#Guardar el dataframe en un archivo csv
df.to_csv("oncologicos.csv",index=False,encoding="latin-1")
```
