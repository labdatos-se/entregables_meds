## Código para obtener listado de enfermedades catastróficas


Este código sirve para obtener la tabla con las enfermedades que causan gastos catastróficos en México. La última actualización de esta es de 2018.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
url="https://www.dof.gob.mx/nota_detalle.php?codigo=5533136&fecha=26/07/2018#gsc.tab=0"
page = requests.get(url, headers=headers,verify=False)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Obtener todos los elementos de la tabla
tabla=soup.find_all('table')
```

```python
#obtener cada uno de los elementos de la tabla
categoria=[]
clave_pt=[]
clave_cie=[]
num_enf=[]
enfermedad=[]
fecha_protocolo=[]
codigo_enf=[]
titulo_de_protocolo_tecnico=[]

for elemento in tabla:
    categoria.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:32.4pt;"))
    clave_pt.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:27.55pt;"))
    clave_cie.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:27.0pt;"))
    num_enf.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:38.64pt;"))
    enfermedad.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:105.8pt;"))
    fecha_protocolo.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:38.2pt;"))
    codigo_enf.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:78.09pt;"))
    titulo_de_protocolo_tecnico.append(elemento.find_all('td',style="border: 0.75pt solid rgb(0,0,0);padding-bottom:0.0pt;padding-left:3.6pt;padding-right:3.6pt;padding-top:0.0pt;vertical-align:middle;width:309.95pt;"))
```

```python
#Obtener texto de cada elemento
for i in range(len(categoria)):
    categoria[i]=[elemento.text for elemento in categoria[i]]
    clave_pt[i]=[elemento.text for elemento in clave_pt[i]]
    clave_cie[i]=[elemento.text for elemento in clave_cie[i]]
    num_enf[i]=[elemento.text for elemento in num_enf[i]]
    enfermedad[i]=[elemento.text for elemento in enfermedad[i]]
    fecha_protocolo[i]=[elemento.text for elemento in fecha_protocolo[i]]
    codigo_enf[i]=[elemento.text for elemento in codigo_enf[i]]
    titulo_de_protocolo_tecnico[i]=[elemento.text for elemento in titulo_de_protocolo_tecnico[i]]
```

```python
#Desplegar todas las filas
pd.set_option('display.max_rows', None)
```

```python
#Hacer dataframe
df=pd.DataFrame({'categoria':categoria[0],'clave_pt':clave_pt[0],'título_de_protocolo_técnico':titulo_de_protocolo_tecnico[0],'clave_cie':clave_cie[0],'num_enf':num_enf[0],'enfermedad':enfermedad[0],'fecha_protocolo':fecha_protocolo[0],'codigo_enf':codigo_enf[0]})
```

```python
#Eliminar espacios adelante y atras en todas las columnas
for columna in df.columns:
    df[columna]=df[columna].str.strip()
```

```python
#Imprimir len con separador de miles
print("Existen {:,} enfermedades catastróficas".format(len(df)),"en México")
df
```
