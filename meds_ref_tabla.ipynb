```python
import pandas as pd
import tabula
```

```python
urlpdf="https://www.gob.mx/cms/uploads/attachment/file/803597/LMR_2023-01_actualizaci_n_10_febrero_2023.pdf"
```

```python
#Leer el pdf
data = tabula.read_pdf(urlpdf, pages='all',lattice=True)
```

```python
#Pasar a dataframe
df=pd.DataFrame()
for tablas in range(len(data)):
    df=pd.concat([df,data[tablas]],axis=0)
#Renombrar columnas
df.columns=['num','farmaco','forma', 'concen','fraccion','den_dist','registro','titular','indicacion']
df.reset_index(drop=True,inplace=True)
```

```python
#Limpieza de datos
#Minúsculas
df=df.apply(lambda x: x.astype(str).str.lower())
#Eliminar espacios en blanco
df=df.apply(lambda x: x.astype(str).str.strip())
#Eliminar acentos
df=df.apply(lambda x: x.astype(str).str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
#Eliminar \r
df=df.apply(lambda x: x.astype(str).str.replace("\r",""))
df['indicacion']=df['indicacion'].str.replace(".","")
df
```

```python
#Salvar a csv
df.to_csv("meds_referencia.csv",index=False)
```
