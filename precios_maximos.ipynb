```python
#Leer pdf
data = tabula.read_pdf("precios_medicamentos_patente.pdf", pages="all",stream=True, columns=False)
```

```python
#Convertir a dataframe sin index
df=pd.DataFrame()
for i in range(len(data)):
    df=pd.concat([df,data[i]],ignore_index=True)

#Dejar las primeras cinco columnas
df=df.iloc[:,0:5]
#Renombrar columnas
df.columns=["empresa","producto","sustancia", "presentacion", "precio_max"]
df
```

```python
#eliminar nans
df=df.dropna()
```

```python
#Eliminar comas en precios
df["precio_max"]=df["precio_max"].str.replace(",","")
#transforamr a float
df["precio_max"]=df["precio_max"].astype(float)
```

```python
df.info()
```

```python
df
```

```python
#Filtrar por medicamentos oncológicos
df=df[df["sustancia"].isin(oncologicos["nombre"])]
```

```python
#datos en minúsculas
df["sustancia"]=df["sustancia"].str.lower()
df["producto"]=df["producto"].str.lower()
#Eliminar espacios en blanco
df["sustancia"]=df["sustancia"].str.strip()
```

```python
#Extrer número antes de la palabra "tabletas". Dejar solo el número
df["num_prod"]=df["presentacion"].str.extract("(\d+ tabletas)")
#Extrer número antes de la palabra "mg". ya sea pegado o con espacio. Dejar solo el número
df["quantity"]=df["presentacion"].str.extract("(\d+mg|\d+ mg)")
#Dejar solo números en ambas columnas
df["quantity"]=df["quantity"].str.extract("(\d+)")
df["num_prod"]=df["num_prod"].str.extract("(\d+)")
#Transformar a float
df["num_prod"]=df["num_prod"].astype(float)
df["quantity"]=df["quantity"].astype(float)
#Eliminar nans
df=df.dropna()
#Renombrar columna sustancia
df=df.rename(columns={"sustancia":"busqueda"})
#Eliminar presentaciones que contengan la palabra "Frasco"
df=df[~df["presentacion"].str.contains("Frasco")]
```

### Cruce con datos obtenidos del webscraping

```python
url_data="https://raw.githubusercontent.com/claudiodanielpc/datos_meds/main/2023-04-09/df_limpia.csv?token=GHSAT0AAAAAACALL3XG472XEOVQ4BXVNTRCZB7BO3Q"
df_limpia=pd.read_csv(url_data,encoding="utf-8")
#Eliminar espacios en product_name
df_limpia["product_name"]=df_limpia["product_name"].str.strip()
#Eliminar registros de país Colombia
df_limpia=df_limpia[df_limpia["pais"]!="Colombia"]
```

```python
#Filtrar df limpia por sustancia de df
df_limpia=df_limpia[df_limpia["busqueda"].isin(df["busqueda"])]
```

```python
df_limpia
```

```python
#Buscar glivec en df_limpia
df_limpia[df_limpia["product_name"].str.contains("glivec")]
```

```python
#Resetear index de df_limpia y df
df_limpia=df_limpia.reset_index(drop=True)
df=df.reset_index(drop=True)
```

```python
#Filtrar df_limpia por busqueda, quantity y num_prod de df
muestra=df_limpia[df_limpia["busqueda"].isin(df["busqueda"]) &
                     df_limpia["quantity"].isin(df["quantity"]) &
                        df_limpia["num_prod"].isin(df["num_prod"])]
```

```python
#Sacar el promedio de precio agrupando por product_name, quantity y num_prod
muestra=muestra.groupby(["busqueda","product_name","quantity","num_prod"]).agg({"precio":"mean"}).reset_index()
```

```python
muestra
```

```python
#Buscar ruxolitinib en df
df[df["busqueda"].str.contains("ruxolitinib")]
```

```python
#cruzar df con muestra de acuerdo a busqueda, quantity y num_prod
cruce=pd.merge(df,muestra[["busqueda","product_name","quantity","num_prod","precio"]],on=["busqueda","quantity","num_prod"],how="left")
#Eliminar nans
cruce=cruce.dropna()
#Eliminar columna product_name
cruce=cruce.drop(columns=["product_name"])
#Precio con dos decimales
cruce["precio"]=cruce["precio"].round(2)
cruce["dif_precio"]=cruce["precio_max"]-cruce["precio"]
#Ordenar dif_precio de mayor a menor
cruce=cruce.sort_values(by="dif_precio",ascending=True)
```
