### Construcción de función para determinar precios promedio


```python
#Hacer la lista de principios
lista_principios=base['principio']
lista_principios=lista_principios.drop_duplicates()
#Reset index
lista_principios=lista_principios.reset_index(drop=True)
#Transformar a dataframe
lista_principios=pd.DataFrame(lista_principios)
lista_principios
```

```python
def prop_prec_mins(tipo_busqueda, busca, excluye_pais="Mexico"):

    if tipo_busqueda not in ['principio', 'atc_5']:
        raise ValueError('la búsqueda solo puede ser por principio activo o atc_5')

    filtro=base[base[tipo_busqueda]==busca]
    #filtro=filtro[filtro["quantity"]==quantity]
    #Excluir país
    if excluye_pais:
        if isinstance(excluye_pais, str):  
            excluye = [excluye_pais]
        filtro=filtro[~filtro["pais"].isin(excluye)]
    #Obtener precio por miligramo o mililitro
    filtro["precio_mg_ml"]=filtro["precio"]/filtro["quantity"]/filtro["num_prod"]
    #Obtener el precio promedio por país y presentación y ponerlo en una tabla
    filtro=filtro.groupby(['pais']).agg({'precio_mg_ml':'mean'}).reset_index()
    #Obtener los 3 precios mínimos y promediarlos
    filtro=filtro.sort_values('precio_mg_ml', ascending=True).head(3)
    #Obtener promedio de promedios internacionales
    prom=filtro["precio_mg_ml"].mean()
    #Crear una nueva base con México y filtrando por principio activo
    mexico = base[(base["pais"] == "Mexico") & (base[tipo_busqueda] == busca)]
    #Dejar las columnas producto, principio activo, presentación, cantidad, precio y precio por miligramo o mililitro
    mexico=mexico[["producto","product_name","farmacia_corto","principio","quantity","unit","num_prod","atc_5","precio"]]
    #Renombrar precio por precio_vigente
    mexico.rename(columns={'precio':'precio_vigente'}, inplace=True)
    #Obtener precio propuesto utilizando prom y multiplicando por cantidad y número de productos
    mexico["precio_propuesto"]=round(mexico["quantity"]*mexico["num_prod"]*prom,2)
    #Eliminar notación científica
    pd.set_option('display.float_format', lambda x: '%.2f' % x)
    #Obtener diferencia entre precio vigente y precio propuesto
    mexico["diferencia"]=round(mexico["precio_vigente"]-mexico["precio_propuesto"],2)
    #Obtener veces
    mexico["veces"]=round(mexico["precio_vigente"]/mexico["precio_propuesto"],2)
    #Ordener de mayor a menor por precio_vigente
    mexico=mexico.sort_values('diferencia', ascending=False)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    
    print("El precio promedio por miligramo o mililitro a nivel internacional es de: ",round(prom,2), "pesos")
    print("El precio promedio por miligramo o mililitro en México es de: ", round((base[(base["pais"] == "Mexico") & (base[tipo_busqueda] == busca)]["precio"]/(base["quantity"]*base["num_prod"])).mean(),2), "pesos")
    print("El precio promedio en México es ", round((base[(base["pais"] == "Mexico") & (base[tipo_busqueda] == busca)]["precio"]/(base["quantity"]*base["num_prod"])).mean()/prom,2), "veces el precio promedio internacional")
    
    return mexico
```
