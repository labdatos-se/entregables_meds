```python
import pandas as pd
import numpy as np
from medscraper import regula
```

```python
pd.set_option('display.max_rows', 10)
```

```python
base=pd.read_excel("C:/Users/claudio.pacheco/Documents/GitHub/entregables_meds/piloto_seguimiento/pa_adic_5sep2023.xlsx")
base.columns=base.columns.str.lower()
#Renombrar columnas principio activo como principio
base.rename(columns={'busqueda':'principio'}, inplace=True)
base
```

```python
base.info()
```

```python
#Hacer la lista de principios
lista_principios=base['principio']
lista_principios=lista_principios.drop_duplicates()
#Reset index
lista_principios=lista_principios.reset_index(drop=True)
#Transformar a dataframe
lista_principios=pd.DataFrame(lista_principios)
lista_principios
```

```python
promedios = pd.DataFrame()
resultados = pd.DataFrame()

for i in range(0, len(lista_principios)):
    result, promedio = regula.regular_precio(base, 'principio', lista_principios.iloc[i, 0], excluye_pais='India')
    
    # Append to the accumulating DataFrames
    resultados = pd.concat([resultados, result])
    promedios = pd.concat([promedios, promedio])

    
```

```python
#Salvar en excel en pestañas separadas
with pd.ExcelWriter('C:/Users/claudio.pacheco/Documents/GitHub/entregables_meds/piloto_seguimiento/precios_regulados.xlsx') as writer:  
    resultados.to_excel(writer, sheet_name='resultados',index=False)
    promedios.to_excel(writer, sheet_name='promedios', index=False)
```
