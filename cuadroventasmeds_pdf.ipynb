```python
import pandas as pd
import tabula as tb
#Para correr el traductor se necesita la versi√≥n 3.1.0a0 de googletrans
from googletrans import Translator
```

```python
file= "https://www.rand.org/content/dam/rand/pubs/research_reports/RR2900/RR2956/RAND_RR2956.pdf"
```

```python
table=tb.read_pdf(file,pages=37, stream=True)
```

```python
df1=pd.DataFrame(table[0])
df1 = pd.concat([df1.columns.to_frame().T, df1], ignore_index=True)
df1.columns = range(len(df1.columns))
df1
```

```python
df0=pd.DataFrame(table[1])
df0
```

```python
#Renombrar columnas del df1 para que coincidan con las del df0
df1.columns = ['Pais', 'Ventas', 'Volumen', 'Part_ Ventas', 'Part Volumen']	  
df0.columns = ['Pais', 'Ventas', 'Volumen', 'Part_ Ventas', 'Part Volumen']
```

```python
df_concatenado = df0.append(df1, ignore_index=True)
```

```python
#Eliminar la primera fila del df_concatenado
df_concatenado = df_concatenado.drop([0], axis=0)
df_concatenado
```

```python
df_concatenado.info()
```

```python
#Utilizar el traductor de google para traducir las columnas del df_concatenado
translator = Translator()
df_esp = df_concatenado.copy()
df_esp["Pais"]=df_esp["Pais"].apply(lambda x: translator.translate(x,dest="es").text)
df_esp
```
