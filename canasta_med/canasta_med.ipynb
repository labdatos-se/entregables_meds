```python
import pandas as pd
from googletrans import Translator
```

```python
#Leer excel
df=pd.read_excel('canasta_med/CNIS_b3_PA.xlsx')
df.columns=df.columns.str.lower()
```

```python
df
```

```python
#Traducir columna pa
translator = Translator()
df['pa_en'] = df['pa'].apply(lambda x: translator.translate(x, src='es', dest='en').text)
#minúsculas
df['pa_en'] = df['pa_en'].str.lower()
```

```python
df
```

```python
##Leer atc
atc=pd.read_excel('atc/atc_clasificador1.xlsx')
atc
```

```python
cruce=df.copy()
#Cruzar con atc
cruce=pd.merge(cruce,atc,how='left',left_on='pa_en',right_on='drug_name')
#Tirar lo que tenga atc5_code nulo
cruce=cruce[cruce['atc5_code'].notnull()]
cruce
```

```python
#Detectar aquellos de df que no lograrón cruzarse
nullc=df[~df['pa_en'].isin(cruce['pa_en'])]
nullc
```

```python
#Guardar en excel en pestañas diferentes
with pd.ExcelWriter('canasta_med/canasta_med.xlsx') as writer:
    df.to_excel(writer, sheet_name='canasta_med',index=False)
    cruce.to_excel(writer, sheet_name='cruce',index=False)
    nullc.to_excel(writer, sheet_name='sin_cruce',index=False)
```
