```python
import pandas as pd
from medscraper.buscadores import busca_cofepris
import time
import random
```

```python
base=pd.read_excel('C:/Users/claudio.pacheco/Documents/GitHub/entregables_meds/canasta_med/imss_col.xlsx')
base.columns=base.columns.str.lower()
base.rename(columns={'principios activos':'principios'},inplace=True)
base['principios']=base['principios'].str.lower()
base
```

```python
#Crear lista con nombre producto por grupos de 50
lista=base['principios'].tolist()
lista2=[lista[i:i + 50] for i in range(0, len(lista), 50)]
lista2
```

```python
busqueda = pd.DataFrame()

for i in lista2:
    for j in i:
        busqueda = pd.concat([busqueda, busca_cofepris.registro(j)])
     # Tiempo entre elementos
        time.sleep(random.randint(1, 2))
    # Tiempo entre listas
    time.sleep(random.randint(120, 240))

busqueda
```

```python
#Cruzar con la base de datos y poner 1 si está en Cofepris y 0 si no
base['cofepris']=base['principios'].isin(busqueda['busqueda']).astype(int)
base
```

```python
#Guarda en excel
with pd.ExcelWriter('imss_col.xlsx') as writer:
    base.to_excel(writer, sheet_name='base', index=False)
    busqueda.to_excel(writer, sheet_name='registros', index=False)
    
```
